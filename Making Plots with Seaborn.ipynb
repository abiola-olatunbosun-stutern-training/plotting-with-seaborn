# Plotting with Seaborn

```python
# Importing Pandas and Seaborn library, loading the dataset, and displaying top 5 rows
import pandas as pd
import seaborn as sns
pokemon = pd.read_csv('Pokemon.csv', index_col=0)
pokemon.head()
```

# Count Plot

```python
# Plotting a countplot chart
sns.countplot(pokemon['Generation'])
```

# Histogram

```python
# Plotting a histogram chart
sns.displot(pokemon['HP'])
```

# Joint Plot (Scatter Plot)

```python
# Plotting a scatterplot chart
sns.jointplot(x='Attack', y='Defense', data=pokemon)
```

# Joint Plot (Hex Plot)

```python
# Plotting a hexplot chart
sns.jointplot(x='Attack', y='Defense', data=pokemon, kind='hex')
```

# KDE Plot

```python
# Plotting a kdeplot chart
sns.kdeplot(pokemon['HP'], pokemon['Attack'])
```

# Box Plot

```python
# Plotting a boxplot chart
sns.boxplot(x='Legendary', y='Attack', data=pokemon)
```

# Violin Plot

```python
# Plotting a violinplot chart
sns.violinplot(x='Legendary', y='Attack', data=pokemon)
```

# Exercises

1. A `seaborn` `countplot` is equivalent to what in `pandas`?
2. A `seaborn` `jointplot` which is configured with `kind='hex'` is equivalent to a what in `pandas`?
3. Why might a `kdeplot` not work very well for ordinal categorical data?
4. What does the "box" in a `boxplot` represent?

# Answers

1. A seaborn countplot is equivalent to a bar plot in pandas.
2. A seaborn jointplot is equivalent to a hex plot in pandas.
3. KDEPlots work by aggregating data into a smooth curve, which is great for interval data, but doesn't always work well for ordinal categorical data.
4. At the top of the box is the 75th percentile, and at the bottom is the 25th percentile. The median, which is the 50th percentile, is the line in the center of the box. This represents that 50% of the data in the distribution is located within the box.
